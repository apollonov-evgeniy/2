###### –ò–∑—É—á–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤

```python
def view_nan(df1): 
    plt.figure(figsize=(12,7))
    # –æ–ø—Ä–µ–¥–µ–ª—è–µ–º —Ü–≤–µ—Ç–∞ 
    # –∂–µ–ª—Ç—ã–π - –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, —Å–∏–Ω–∏–π - –Ω–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ
    colours = ['#000099', '#ffff00'] 
    sns.heatmap(df1[list(df1)].isnull(), cmap=sns.color_palette(colours))
    # –≤—ã–≤–µ–¥–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç—ã –ø—Ä–æ–ø—É—Å–∫–æ–≤ 
    for col in df1.columns:
        pct_missing = np.mean(df1[col].isnull())
        pct_missing_sum = df1[col].isnull().sum()
        if pct_missing_sum > 0:
            print(f'{pct_missing_sum} –ø—Ä–æ–ø—É—Å–∫–∞ –≤ {col} -  —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç -{pct_missing:.2%}')
```

–°—Ç–æ–ª–±–µ—Ü year_of_release (–≥–æ–¥ –≤—ã–ø—É—Å–∫–∞ –∏–≥—Ä—ã) –∏–º–µ–µ—Ç 1,6% –ø—Ä–æ–ø—É—Å–∫–æ–≤. –°—Ä–µ–¥–Ω–∏–º–∏/–º–µ–¥–∏–∞–Ω–∞–º–∏ —ç—Ç–∏ –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–µ –∑–∞–ø–æ–ª–Ω–∏—Ç—å. –ó–∞–º–µ–Ω–∏–º –Ω–∞ –Ω–µ —Ä–µ–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ - –ø—Ä–∏–∑–Ω–∞–∫ 1900. (–í –ø–æ—Å–ª–µ–¥—Å—Ç–≤–∏–∏ –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ –∏–∑—É—á–µ–Ω–∏—è, –±—ã–ª–æ –ø—Ä–∏–Ω—è—Ç–æ —Ä–µ—à–µ–Ω–∏–µ –∑–∞–º–µ–Ω–∏—Ç—å –º–µ–¥–∏–∞–Ω–æ–π —Å –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–æ–π –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º) 

–ü—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö name –∏ genry –≤ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —Å—Ç—Ä–æ–∫–∞—Ö(–≤—Å–µ–≥–æ 2) - —É–¥–∞–ª–∏–º

###### –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ –∑–Ω–∞—á–µ–Ω–∏—è 'tbd', –ò–∑–º–µ–Ω–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –Ω–µ—è–≤–Ω—ã–µ –ø—Ä–æ–ø—É—Å–∫–∏ `tbd` –≤ NaN
games.loc[games.user_score == 'tbd' , 'user_score'] = np.NaN
```

```python
# –ó–∞–º–µ–Ω–∏–º –∑–Ω–∞—á–µ–Ω–∏—è –∫–∞–∫ –æ–ø–∏—Å–∞–Ω–æ —Ä–∞–Ω–µ–µ –∏ –ø–æ–º–µ–Ω—è–µ–º —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö
# user_score –∏–∑ —Å—Ç—Ä–æ–∫–æ–≤–æ–≥–æ (–ø–æ—Å–ª–µ –∑–∞–º–µ–Ω—ã *tbd*) –≤ float
# critic_score float --> int
games.user_score = games.user_score.fillna(-1).astype('float')
games.critic_score = games.critic_score.fillna(-1).astype('int')
games.rating = games.rating.fillna('No')
```

```python
# –ü–æ—Å–ª–µ –∏–∑—É—á–µ–Ω–∏—è "–∂–∏–∑–Ω–∏" –ø–ª–∞—Ç—Ñ–æ—Ä–º, –±—ã–ª–æ –ø—Ä–∏–Ω—è—Ç–æ —Ä–µ—à–µ–Ω–∏–µ –∑–∞–º–µ–Ω–∏—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –≤ –≥–æ–¥–∞—Ö
# –Ω–∞ –º–µ–¥–∏–∞–Ω—É –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º. –ó–∞–º–µ–Ω–∞ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫ (1900 –¥–∞–ª—å—à–µ –Ω–µ –Ω—É–∂–Ω–∞. –≤ –∫–æ–¥–µ –æ—Å—Ç–∞–ª–∞—Å—å)
year_of_release_median =(
    games
    .groupby(['platform'])
    .agg({'year_of_release': 'median'})['year_of_release']
    .round()
    .to_dict()
)

(games
    .year_of_release
    .fillna(games.loc[games.year_of_release.isnull(),'platform']
                  .map(year_of_release_median), inplace=True))


# –≥–æ–¥ —Ä–µ–ª–∏–∑–∞ NaN –Ω–∞ 1900, 
#float--> int, —Ç.–∫. –≥–æ–¥ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ
games.year_of_release = games.year_of_release.fillna(1900)
games.year_of_release = games.year_of_release.astype('int')
 
```

```python
# —É–¥–∞–ª—è–µ–º —Å—Ç—Ä–æ—á–∫–∏, –≥–¥–µ name –∏ genry - NaN
games=(games[~games.name.isnull()])
games=(games[~games.genre.isnull()])
```

–ë—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å t-–∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞, —Ç–∞–∫ –∫–∞–∫ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º—ã–µ –≤—ã–±–æ—Ä–∫–∏ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã –º–µ–∂–¥—É —Å–æ–±–æ–π.–û–ø—Ä–µ–¥–µ–ª–∏–º –ø–æ—Ä–æ–≥–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ alpha = 0.05

###### –ì–∏–ø–æ—Ç–µ–∑—ã 1:
H0: —Å—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π  
H1: —Å—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è

<div class="alert alert-success">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b>

–ì–∏–ø–æ—Ç–µ–∑—ã —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≤–µ—Ä–Ω–æ.
    
    
**Best practies:**
    
–ü–µ—Ä–µ–¥ –ø—Ä–æ–≤–µ—Ä–∫–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã —Ä–µ–∫–æ–º–µ–Ω–¥—É—é –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫, —á—Ç–æ–±—ã –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∞—Ç—Ä–∏–±—É—Ç `equal_val` –≤ –¢‚Äì—Ç–µ—Å—Ç–µ.
    
–í —Ç–æ–º —á–∏—Å–ª–µ —Å—Ç–æ–∏—Ç –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–∞–∂–¥–æ–π –≤—ã–±–æ—Ä–∫–∏. 
    
</div>

```python
# —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å—Ç–∞—Ç –ø–∞—Ä–∞–º–µ—Ä–æ–≤
def compare_stat(df1, df2, column_df):
    df1_mean=df1[column_df].mean()
    df2_mean=df2[column_df].mean()
    # –≤—ã–±–æ—Ä–æ—á–Ω–∞—è —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç—å ddof=1
    df1_variance =np.var(df1[column_df], ddof=1)
    df2_variance =np.var(df2[column_df], ddof=1)
    
    df1_standard= np.std(df1[column_df], ddof=1)
    df2_standard= np.std(df2[column_df], ddof=1)
    
    print(f'Smart –ø–æ ({column_df}) –°—Ä–µ–¥–Ω–µ–µ - {df1_mean:.2f}, \
        –î–∏—Å–ø–µ—Ä—Å–∏—è - {df1_variance:.2f}, –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ - {df1_standard:.2f}')
    print(f'Ultra –ø–æ ({column_df}) –°—Ä–µ–¥–Ω–µ–µ - {df2_mean:.2f}, \
        –î–∏—Å–ø–µ—Ä—Å–∏—è - {df2_variance:.2f}, –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ - {df2_standard:.2f}')
```

```python
# –≤—Å–ø–æ–º–∏–Ω–∞–µ–º, —á—Ç–æ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞ –∑–∞–º–µ–Ω–µ–Ω–∏–Ω—ã NaN –Ω–∞ -1


alpha = 0.05 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, 
#–µ—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –º–µ–Ω—å—à–µ –µ–≥–æ - –æ—Ç–≤–µ—Ä–≥–Ω–µ–º –≥–∏–ø–æ—Ç–µ–∑—É
xone = games.query('platform == "xone" and user_score  >= 0').user_score
pc = games.query('platform == "pc" and user_score  >= 0').user_score
```

–í—ã–¥–≤–∏–Ω–µ–º –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ, —á—Ç–æ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ —Ä–∞–∑–ª–∏—á–∏–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞ –¥–ª—è –¥–≤—É—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ–ª—É—á–µ–Ω—ã —Å–ª—É—á–∞–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º –∏ –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –≤ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏  –Ω–µ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è  
(H0: —Å—Ä–µ–¥–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π)

```python
result=(
    st
    .ttest_ind(xone, pc)
)
print(f'p-–∑–Ω–∞—á–µ–Ω–∏–µ: {result.pvalue}')

if result.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
    
```
